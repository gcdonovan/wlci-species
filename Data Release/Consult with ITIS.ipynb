Consult the ITIS API to retrieve information and cache for later processing. This is done via the ITIS module in the bispy package that uses various search mechanisms against ITIS and slightly repackages the returned data for use.

```python
import requests
import json
import bispy
from IPython.display import display
from joblib import Parallel, delayed
import random
import pandas as pd

itis = bispy.itis.Itis()

import warnings
warnings.filterwarnings('ignore')
```

```python
# Open source WLCI list created from build-specie-list.ipynb
with open("sources/wlci_xdd_specie_list.txt", "r") as f:
    sp_list =f.readlines()
sp_list= [x.strip() for x in sp_list]
```

```python
#print initial list
sp_list
```

```python
# Use joblib to run multiple requests for ITIS documents in parallel via known ITIS TSNs
itis_cache = Parallel(n_jobs=8)(delayed(itis.search)(name_or_tsn) for name_or_tsn in sp_list)
```

```python
# Dump the ITIS data to a cache file
with open("cache/itis.json", "w") as f:
    f.write(json.dumps(itis_cache, indent=4))
```

```python
# Open the file back up and verify
with open("cache/itis.json", "r") as f:
    itis_cache = json.loads(f.read())

print(len(itis_cache))
display(itis_cache[random.randint(0,len(itis_cache)-1)])
```
